```python
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt

# Load the image into a numpy array
img = Image.open("test.png")
img_array = np.array(img)
# img_array = (y,x)
new_array_l = []
new_array_r = []
starting_x = 1313
height = img_array.shape[0] - 1
points = []
detected_points = []

for y in range(height, -1, -1):  # range(height, -1, -1)
    new_x_l = starting_x
    new_x_r = starting_x
    while img_array[y, new_x_l - 1].all() == img_array[y, new_x_l].all():
        new_x_l -= 1

    while img_array[y, new_x_r + 1].all() == img_array[y, new_x_r].all():
        new_x_r += 1

    new_array_l.append(new_x_l)
    new_array_r.append(new_x_r)


# print(new_array)
```
